```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
df.drop('custid',axis="columns",inplace=True)
```

```python
df.drop('Unnamed: 8',axis="columns",inplace=True)
df.drop('Unnamed: 9',axis="columns",inplace=True)
df.drop('Unnamed: 10',axis="columns",inplace=True)
df.drop('Unnamed: 11',axis="columns",inplace=True)
df.drop('Unnamed: 12',axis="columns",inplace=True)
df.drop('Unnamed: 13',axis="columns",inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2= df
df2.favday = le.fit_transform(df2.favday)
df2
```

```python
df2.isna().sum()
```

```python
y=df2['retained']
x=df2.drop(['retained'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC

svc=SVC(C= .1, kernel='linear', gamma= 1)
svc.fit(X_train,y_train)

xbg_model = XGBClassifier()
xbg_model.fit(X_train, y_train)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

model=LogisticRegression()
model.fit(X_train,y_train)

print('KNN:{:.2f}'.format(knn.score(X_test,y_test)*100))
print('RANDOM_FOREST:{:.2f}'.format(rf.score(X_test,y_test)*100))
print('LOG_REG:{:.2f}'.format(model.score(X_test,y_test)*100))
print('XGB:{:.2f}'.format(xbg_model.score(X_test,y_test)*100))
print('SVM:{:.2f}'.format(svc.score(X_test,y_test)*100))

from sklearn.ensemble import VotingClassifier
estimators=[('knn', knn), ('rf', rf), ('log_reg', model),('XGB',xbg_model),('SVM',svc)]
ensemble = VotingClassifier(estimators,voting='hard')


ensemble.fit(X_train, y_train)
final_score=ensemble.score(X_test,y_test)

print(final_score*100, '%')
```
